```python
import pandas as pd
import os
import operator
```

```python
os.chdir('C:\\tmp')
os.getcwd()

data = pd.read_csv('Taller1Datos.csv',sep=';',header=0,engine='python')
data.head()
```

```python
data.rename(columns={'¿Por qué estás estudiando análitica estratégica de datos?': 'Pq_Analitica'}, inplace=True)
```

```python
data['Qpalabras']=data['Pq_Analitica'].str.split().str.len()
data
```

```python
data=data.dropna(how="all")
data
```

```python
FramePalabras = data['Pq_Analitica'].str.split()
```

```python
listaPalabras=[]
for k, v in FramePalabras.items():
    for i in v:
        listaPalabras.append(i)
listaPalabras
```

```python
frecuencia = []
for palabra in listaPalabras:
    frecuencia.append(listaPalabras.count(palabra))
frecuencia        
```

```python
Q=list(zip(listaPalabras, frecuencia))
Q = sorted(Q, key=operator.itemgetter(1), reverse=True)
Q[0][0]
print("La palabra que mas se repite es: " + str(Q[0][0]))
print("Se repite : " + str(Q[0][1])+" Veces")
```
